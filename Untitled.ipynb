```python
# === Data cleaning & preprocessing ===
import pandas as pd
import numpy as np

# Adjust filename/path if you renamed the file
csv_path = 'data/WA_Fn-UseC_-Telco-Customer-Churn.csv'
df = pd.read_csv(csv_path)
#number of rows and columns
print("Initial shape:", df.shape)
#list of all columns
print("Columns:", df.columns.tolist())
# Check for missing values NaN
df.isnull().sum()
## Check data types
df.dtypes
#convert TotalCharges to numeric
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
# Count rows with NaN TotalCharges
n_missing_total = df['TotalCharges'].isna().sum()
print("\nRows with non-numeric TotalCharges:", n_missing_total)
# Drop rows where TotalCharges is NaN 
df = df[~df['TotalCharges'].isna()].copy()
print("Shape after dropping TotalCharges NaNs:", df.shape)
# -- 4) Standardize common binary columns (map Yes/No to 1/0)
binary_cols = []
for col in df.columns:
    # detect columns that are object and have only Yes/No values (plus maybe NaN)
    if df[col].dropna().unique().tolist() and df[col].dtype == 'object':
        unique_vals = set(df[col].dropna().unique())
        if unique_vals <= {'Yes','No'}:
            binary_cols.append(col)

print("\nDetected binary (Yes/No) columns:", binary_cols)
for col in binary_cols:
    df[col] = df[col].map({'Yes':1, 'No':0}).astype(int)
# Encode Churn specifically (if not already numeric)
if 'Churn' in df.columns and df['Churn'].dtype == 'object':
    df['Churn'] = df['Churn'].map({'Yes':1, 'No':0}).astype(int)
n_dup = df.duplicated().sum()
print("\nDuplicate rows found:", n_dup)
if n_dup > 0:
    df = df.drop_duplicates().copy()
    print("Shape after dropping duplicates:", df.shape)

# -- 6) Final quick checks
print("\nFinal shape:", df.shape)
print("\nData types:\n")
display(df.dtypes)

print("\nMissing values per column:")
display(df.isna().sum())

print("\nChurn distribution (counts & percent):")
display(df['Churn'].value_counts())
display((df['Churn'].value_counts(normalize=True)*100).round(2))
# Save cleaned copy (optional)
df.to_csv('data/telco_cleaned.csv', index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(data = df, x = "Churn")
plt.title("Churn Distribution")
plt.show()
```

The dataset shows that approximately 27% of customers churn.
This is a significant portion of the customer base, indicating a churn problem that needs attention.

```python

plt.figure(figsize=(8,6))
sns.barplot(
    data=df ,
    x='Contract',
    y='Churn',
    estimator='mean',
    errorbar=None
)
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate')
plt.xlabel('Contract Type')
plt.xticks(rotation=20)
plt.ylim(0, 0.6)
plt.show()
```

Customers on Month-to-Month contracts churn dramatically more than One-year or Two-year contracts.
Recommendation: Offer incentives to switch them to 1-year or 2-year contracts.

```python
plt.figure(figsize=(8,6))
sns.boxplot(
    data=df,
    x='Churn',
    y='MonthlyCharges',
)

plt.title('Monthly Charges vs Churn')
plt.xlabel('Churn (0 = No, 1 = Yes)')
plt.ylabel('Monthly Charges')
plt.show()
```

Churners have a higher median MonthlyCharge → price-sensitive customers are at risk.

```python
plt.figure(figsize=(8,6))
sns.kdeplot(
    data=df,
    x='tenure',
    hue='Churn',
    common_norm=False,
    fill=True,
    alpha=0.4
)

plt.title('Tenure Distribution by Churn')
plt.xlabel('Tenure (months)')
plt.ylabel('Density')
plt.show()
```

Insight: Customers with short tenure churn early.
The density curve for churners is heavily concentrated in tenure < 12 months.
Recommendation: Implement welcome programs or early engagement strategies.

```python
sns.countplot(data=df, x="SeniorCitizen", hue="Churn")
plt.title("Churn by Senior Citizen Status")
plt.show()
```

Insight: Non-seniors are more likely to stay, seniors churn rate is about equal.
Recommendation: Consider targeted retention strategies for senior customers.
